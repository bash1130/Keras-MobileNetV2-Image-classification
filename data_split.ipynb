```python
import os
rootDir='/home/eric/data/scene'
file1=os.path.join(rootDir,'scene1.txt')
file2=os.path.join(rootDir,'scene2.txt')
file3=os.path.join(rootDir,'scene3.txt')
```

```python
list_files=[]
with open(file1,'r') as f:
    arr=f.readlines()
    list_files.extend(arr)
with open(file2,'r') as f:
    arr=f.readlines()
    list_files.extend(arr)
with open(file3,'r') as f:
    arr=f.readlines()
    list_files.extend(arr)
```

```python
list_clean=[]
print(list_files[0].strip().split('.'))
for item in list_files:
  
    if(item.strip().split('.')[-1]=='jpg'):
        list_clean.append(item)
    else:
        print(item)
    
```

```python
import random
print(list_clean[:5])
random.shuffle(list_clean)
print(list_clean[:5])
print(len(list_clean))
```

```python
def div_list(ls,n):
    result = []
    cut = int(len(ls)/n)
    if cut == 0:
        ls = [[x] for x in ls]
        none_array = [[] for i in range(0, n-len(ls))]
        return ls+none_array
    for i in range(0, n-1):
        result.append(ls[cut*i:cut*(1+i)])
    result.append(ls[cut*(n-1):len(ls)])
    return result
```

```python
res=div_list(list_clean,10)
print(len(res[0]))
print(len(res[1]))
```

```python
group=os.path.join(rootDir,'group')
for i in range(len(res)):
    group_name=group+str(i)+'.txt'
    with open(group_name,'w') as f:
        f.write('image,label\n')
        for item in res[i]:
            label=item.strip().split('/')[-2]
            f.write(item.strip()+','+label+'\n')
```

```python
for i in range(10):
    group_name=group+str(i)+'.txt'
    with open(group_name,'r') as f:
        arr=f.readlines()
        arr_size=len(arr)
        train_size=int(arr_size*0.6)
        val_size=int(arr_size*0.2)
        group_train=group+str(i)+'_train.txt'
        with open(group_train,'w') as f1:
            for j in range(train_size):
                f1.write(arr[j])
        group_valid=group+str(i)+'_valid.txt'
        with open(group_valid,'w') as f2:
            for j in range(train_size,train_size+val_size):
                f2.write(arr[j])
        group_test=group+str(i)+'_test.txt'
        with open(group_test,'w') as f3:
            for j in range(train_size+val_size,arr_size):
                f3.write(arr[j])
            
```
